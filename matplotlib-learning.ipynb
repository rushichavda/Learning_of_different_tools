```python
x = [0,1,2,3,4,]
y = [0,2,4,6,8]

#resize your Graph(pdi specify pixels per inch .when saving probably should use 300 if possible)
plt.figure(figsize=(5,3),dpi=300)

#plt.plot(x,y,label="2x",color='red',linewidth=2,marker =".",linestyle='--', markersize=10,markeredgecolor="blue")

#Use shorthand notation
#fmt ="[color][marker][Line]"
plt.plot(x,y,'b^-.',label="2x")

#line2

#select interval we want to plot points at
x2 = np.arange(0,4,0.5)

#plot part of the graph as line
plt.plot(x2[:5],x2[:5]**2,'r',label='x^2')

#plot part of the line as dot
plt.plot(x2[4:],x2[4:]**2,'r--',label='x^2')


#add a title (specify font parameter with fontdict) 
plt.title("Our first graph!",fontdict ={'fontname':"Comic Sans MS",'fontsize': 20})

#X and Y label
plt.xlabel("xaxis")
plt.ylabel("yaxis")


#X and Y tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8])

#Add a lagend
plt.legend()

#save figure(dpi 3000 is good when saving so graph has high resolution) 
plt.savefig('my_graph.png',dpi = 300)

# Show plot
plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv("../input/FIFA-data/data.csv")
fifa
```

### Histograms

```python

bins = np.arange(40,110,10)

plt.hist(fifa.Overall, bins=bins,color='#ed701c')
plt.xticks(bins)
plt.ylabel("Number of Players")
plt.xlabel("Skill Level")
plt.title("Distribution of Player skills in FIFA 2018")
plt.show()
```

### Pie Chart

```python
left = fifa.loc[fifa["Preferred Foot"]=="Left"].count()[0]
right = fifa.loc[fifa["Preferred Foot"]=="Right"].count()[0]

labels = ["left","Right"]
colors = ["#abcdef","#ed701c"]


plt.pie([left,right],labels=labels,colors = colors,autopct = '%.2f %%')
plt.title("Foot Preference of FIFA Player")
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot') #to change from default color design

light = fifa.loc[fifa.Weight<125].count()[0]
light_medium = fifa.loc[(fifa.Weight>=125) & (fifa.Weight<150)].count()[0]
medium = fifa.loc[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight>=175) & (fifa.Weight<200)].count()[0]
heavy = fifa.loc[(fifa.Weight>=200)].count()[0]

weights = [light, light_medium, medium , medium_heavy, heavy]
labels =["Under 125","125-150","150-175","175-200","over 200"]
explode = (0.4,0.2,0,0,0.4)

plt.title("Weight Distribution of FIFA Players (in lbs)")

plt.pie(weights,labels = labels,autopct = '%.2f %%',pctdistance = 0.8,explode= explode)

plt.show() 
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid','NE Revolution']

boxes = plt.boxplot([barcelona,madrid,revs],labels = labels,patch_artist=True,medianprops={'linewidth':2})
#medianprops is to change the properties of the median line

for box in boxes['boxes']:
    
    box.set(color='#4286f4',linewidth = 2)# to set the color of box outside periferi(edge)
    #Change fill color
    box.set(facecolor='#e0e0e0')#only change color if the patch_artist is set to
                                #True otherwise it will through an error
    

plt.title ('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.savefig("FIFA Football.png",dpi=300)
plt.show()
 
```
